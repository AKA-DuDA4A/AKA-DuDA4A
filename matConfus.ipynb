```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt


y_real = [1, 0, 1, 0, 1, 1, 0, 0, 1, 0]
y_pred = [1, 0, 0, 0, 1, 1, 1, 0, 1, 1]

mc = confusion_matrix(y_real, y_pred)

plt.figure(figsize=(4,2))
sns.heatmap(mc, annot=True, fmt='d', cmap='Greens', cbar=False)
plt.xlabel('Preditos')
plt.ylabel('Reais')
plt.title('Confusion Matrix')
plt.show()

tp = mc[1,1]
tn = mc[0,0]
fp = mc[0,1]
fn = mc[1,0]

precision = tp / (tp + fp)
recall = tp / (tp + fn)

f1_score = 2 * (precision * recall) / (precision + recall)


print(mc)
print(f'Precisão: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1_score:.2f}')
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score


np.random.seed(42)
y_reais = np.random.randint(0, 2, 200)
y_pred = np.random.randint(0, 2, 200)


mc = confusion_matrix(y_reais, y_pred)

acuracia = accuracy_score(y_reais, y_pred)
precisao = precision_score(y_reais, y_pred)
sensibilidade = recall_score(y_reais, y_pred)
f1 = f1_score(y_reais, y_pred)


plt.figure(figsize=(14, 6))


plt.subplot(1, 2, 1)
sns.heatmap(mc, annot=True, fmt='d', cmap='Greens',
            xticklabels=['Previsto: 0', 'Previsto: 1'],
            yticklabels=['Real: 0', 'Real: 1'],
            annot_kws={"size": 16})
plt.title('Matriz de Confusão - Diagnóstico', fontsize=16, fontweight='bold')
plt.xlabel('O que o modelo previu', fontsize=12)
plt.ylabel('O que era na realidade', fontsize=12)


plt.subplot(1, 2, 2)
metricas_nomes = ['Acurácia', 'Precisão', 'Sensibilidade', 'F1 Score']
metricas_valores = [acuracia, precisao, sensibilidade, f1]
cores = ['#94a3b8', '#38bdf8', '#22c55e', '#facc15']

barras = plt.bar(metricas_nomes, metricas_valores, color=cores)
plt.title('Métricas de Desempenho', fontsize=16, fontweight='bold')
plt.ylim(0, 1.1)


for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2., altura + 0.02,
             f'{altura*100:.1f}%', ha='center', va='bottom', fontweight='bold', fontsize=12)


plt.tight_layout()
plt.show()
```
